```python
#Q3)Explain the following terms
#The difference in np.array, np.asarray and np.asanyarray
#The difference between Deep copy and shallow copy

#1)np.array:

#Purpose: Creates a new array.
#Behavior: Always creates a new array object. It doesn't check if the input is already an array, so even if you pass an existing array, a new array (copy) is created.
#Use case: When you explicitly want to create a new array, regardless of the input type.
import numpy as np

a = [1, 2, 3]
b = np.array(a)
c = np.array(b)  # Creates a new array, even though b is already an array

```

```python
#np.asarray:

#Purpose: Converts the input to an array.
#Behavior: If the input is already an array (ndarray), it doesn't create a new array; it just returns the input array. If the input is not an array, it converts it to an array.
#Use case: When you want to ensure you have an array, but avoid unnecessary duplication if the input is already an array.
a = [1, 2, 3]
b = np.asarray(a)
c = np.asarray(b)  # Returns b itself, no new array created
```

```python
#np.asanyarray:

#Purpose: Similar to np.asarray, but it converts the input to an array or a subclass of an array (e.g., matrix, masked array).
#Behavior: If the input is already an array or an array subclass, it returns the input as-is. If the input is not an array, it converts it to an array (or a subclass if specified).
#Use case: When you want to ensure you have an array-like object, but want to preserve subclasses of arrays.

a = [1, 2, 3]
b = np.asanyarray(a)
c = np.asanyarray(b)  # Returns b itself, no new array created
```

```python
#Deep Copy:

#Definition: A deep copy creates a new object and recursively copies all nested objects, so the new object is completely independent of the original.
#Behavior: Changes to the nested objects in the copied object won't affect the original object, and vice versa.
#Use case: When you need a completely independent copy of a collection, including all nested objects.
import copy
original_list = [[1, 2, 3], [4, 5, 6]]
deep_copy = copy.deepcopy(original_list)

deep_copy[0][0] = 100
# Only deep_copy will show the change; original_list remains unchanged
```

```python
#Shallow Copy:

#Definition: A shallow copy creates a new object, but doesn't create copies of nested objects. Instead, it just copies the references to the nested objects.
#Behavior: Changes to the nested objects will reflect in both the original and the copied object because they share the same references.
#Use case: When you want to create a new collection object (e.g., list, dictionary), but it's acceptable (or desired) that nested objects remain shared.

import copy

original_list = [[1, 2, 3], [4, 5, 6]]
shallow_copy = copy.copy(original_list)

shallow_copy[0][0] = 100
# Both original_list and shallow_copy will show the change
```

```python
#Q14)Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib: right y-axis for 'Humidity'). b) Label the x-axis as 'Date'. a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and c) Set the title of the plot as 'Temperature and Humidity Over Time'.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Generate random temperature and humidity values
temperature = np.random.uniform(low=20, high=35, size=100)  # Random temperatures between 20 and 35
humidity = np.random.uniform(low=30, high=90, size=100)     # Random humidity between 30 and 90

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plot the data
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create another y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
fig.tight_layout()  # Adjust layout to make room for the title
plt.show()
```

```python
#Q15)Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.
#b) Overlay a line plot representing the normal distribution's probability density function (PDF).
#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'. #
#d) Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.randn(1000)

# Plot the histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Calculate the mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Generate values for the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)

# Plot the PDF
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Q17)Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})

# Define a function to assign quadrant labels
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Quadrant I'
    elif x < 0 and y >= 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Apply function to assign quadrants
df['quadrant'] = df.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='viridis')

# Add a legend
plt.legend(title='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
#Q18)With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh output in Jupyter notebooks
output_notebook()

# Generate data
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis', width=800, height=400)

# Add a line renderer
p.line(x, y, line_width=2, color='blue', legend_label='Sine wave')

# Add grid lines
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)

```

```python
#Q19)Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd
import numpy as np

# Enable Bokeh output in Jupyter notebooks
output_notebook()

# Generate random categorical data
categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a Bokeh figure
p = figure(x_range=df['categories'], title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           width=800, height=400, toolbar_location=None)

# Add a bar renderer with color based on value
p.vbar(x='categories', top='values', width=0.5, source=source,
       line_color='white', fill_color='blue')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Customize plot appearance
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
#Q20)Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

import numpy as np
import plotly.graph_objects  as go
import plotly.express as px
# Generate random data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.random.randn(100).cumsum()

# Create a Plotly figure
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout with title and axis labels
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
```
